```python
df['Gender']=df['Gender'].replace('Fe Male','Female')
df['MaritalStatus']=df['MaritalStatus'].replace('Single','Unmarried')
```

```python
df.head()
```

```python
##check missing values
##these are the features with nan values
features_with_na=[features for features in df.columns if df[features].isnull().sum()>=1]
for feature in features_with_na:
    print(feature,np.round(df[feature].isnull().mean()*100,5), '% missing values')
```

```python
df[features_with_na].select_dtypes(exclude='object').describe()
```

```python
#age
df.Age.fillna(df.Age.median(),inplace=True)

#TypeofContact
df.TypeofContact.fillna(df.TypeofContact.mode()[0],inplace=True)

#DurationofPitch
df.	DurationOfPitch.fillna(df.	DurationOfPitch.median(),inplace=True)

#NumberofFollowups
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0],inplace=True)

#PreferredPropertyStar
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0],inplace=True)

#NumberofTrips
df.NumberOfTrips.fillna(df.NumberOfTrips.median(),inplace=True)

#NumberofChildrenVisiting
df.NumberOfChildrenVisiting	.fillna(df.NumberOfChildrenVisiting.mode()[0],inplace=True)

#MonthlyIncome
df.MonthlyIncome.fillna(df.MonthlyIncome.median(),inplace=True)
```

```python
df.head()
df.isnull().sum()
```

```python
df.drop('CustomerID',inplace=True,axis=1)
```

```python
##feature enginering
df['TotalVisiting']=df['NumberOfPersonVisiting']+ df['NumberOfChildrenVisiting']
df.drop(columns=['NumberOfPersonVisiting','NumberOfChildrenVisiting'])
```

```python
##get all the numeric features
num_features=[features for features in df.columns if df[features].dtype!='O']
print('Num of Numerical Features:' ,len(num_features))
```

```python
##get all the numeric features
cat_features=[features for features in df.columns if df[features].dtype=='O']
print('Num of Categorical Features:' ,len(cat_features))
```

```python
##discrete feautres
discrete_feautres=[features for features in num_features if len(df[features].unique())<=25]
print('Num of Discrete features:',len(discrete_feautres))
```

```python
##continous feautres
continous_feautres=[features for features in num_features if features not in discrete_feautres]
print('Num of continous features:',len(continous_feautres))
```

```python
##train test split and model training
from sklearn.model_selection import train_test_split
X=df.drop(['ProdTaken'],axis=1)
y=df['ProdTaken']
```

```python
y.value_counts()
```

```python
X.head()
```

```python
##train test splits
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=25)
X_train.shape,X_test.shape
```

```python
X.info
```

```python
## create column tranformer with 3 types of tranformers
cat_feautres=X.select_dtypes(include='object').columns
num_feautres=X.select_dtypes(exclude='object').columns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(drop='first')

preprocessor=ColumnTransformer(
    [
    ('OneHotEncoder',oh_transformer,cat_feautres),
    ('StandardScaler',numeric_transformer,num_feautres)
    ]
)
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
##apply transformation on test(transform)
X_test=preprocessor.fit_transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
##random forest classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.treimport DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score
```

```python
models={
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier() 
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)#train model

    ##make predictions
    y_train_pred=model.predict(X_train)
    y_test_pred=model.predict(X_test)
    
    ##training set performance
    model_train_accuracy=accuracy_score(y_train,y_train_pred)
    model_train_f1=f1_score(y_train,y_train_pred,average='weighted')
    model_train_precision=precision_score(y_train,y_train_pred)
    model_train_recall=recall_score(y_train,y_train_pred)
    model_train_roc_auc_score=roc_auc_score(y_train,y_train_pred)

     ##training set performance
    model_test_accuracy=accuracy_score(y_test,y_test_pred)
    model_test_f1=f1_score(y_test,y_test_pred,average='weighted')
    model_test_precision=precision_score(y_test,y_test_pred)
    model_test_recall=recall_score(y_test,y_test_pred)
    model_test_roc_auc_score=roc_auc_score(y_test,y_test_pred)

    print(list(models.keys()))

    print("Model performance for training set")
    print("-Accuracy {:.4f}".format(model_train_accuracy))
    print("-Precision  {:.4f}".format(model_train_f1) )
    print("-recall socre {:.4f}".format(model_train_recall) )
    print("-roc auc score {:.4f}".format(model_train_roc_auc_score) )

    print("Model performance for testing set")
    print("-Accuracy {:.4f}".format(model_test_accuracy))
    print("-Precision  {:.4f}".format(model_test_f1) )
    print("-recall socre {:.4f}".format(model_test_recall) )
    print("-roc auc score {:.4f}".format(model_test_roc_auc_score) )
    
    
```

```python
rf_params={"max_depth":[5,8,15,None,10],
           "max_feautres":[5,7,'auto',8],
           "min_samples":[2,8,15,20],
           "n_estimators":[100,200,500,1000]
}
```

```python
randomcv_models=[
    ("RF",RandomForestClassifier(),rf_params),
]
```

```python
rf_params
```

```python
from sklearn.model_selection import RandomizedSearchCV
model_param={}
for name,model,params in randomcv_models:
    random=RandomizedSearchCV(estimator=model,param_distributions=params,
                               n_iter=100,
                               cv=3,
                               verbose=2,
                              n_jobs=-1)
    random.fit(X_train,y_train)
    model_param[name]=random.best_params_

for model_name in model_param:
    print(f"------Best params for {model_name}-----")
    print(model_param[model_name])
```
